```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("comparisons.csv")
df2 = pd.read_csv("assignments.csv")
df3 = pd.read_csv("timetaken.csv")
```

```python
plt.figure(figsize=(12, 6))
# ---------- COMPARISONS ----------
plt.plot(df1["n"], df1["avg_comparisons_both"], marker="s", linestyle="-", label="Avg comparisons (AgeName)")
plt.plot(df1["n"], df1["avg_comparisons_age"], marker="o", linestyle="--", label="Avg comparisons (Age-only)")
plt.plot(df1["n"], df1["avg_comparisons_name"], marker="^", linestyle="--", label="Avg comparisons (Name-only)")
# ---------- ASSIGNMENTS ----------
plt.plot(df2["n"], df2["avg_assignments_both"], marker="D", linestyle="-", label="Avg assignments (AgeName)")
plt.plot(df2["n"], df2["avg_assignments_age"], marker="v", linestyle="--", label="Avg assignments (Age-only)")
plt.plot(df2["n"], df2["avg_assignments_name"], marker=">", linestyle="--", label="Avg assignments (Name-only)")

plt.xticks(range(10, 101, 10))
plt.xlabel("n (records per dataset)")
plt.ylabel("Average comparisons / assignments")
plt.title("Insertion Sort  Avg Comparisons & Assignments vs n")
plt.grid()
plt.legend()

plt.savefig("average_plot.png", dpi=200)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df3["n"], df3["average_time(us)"], marker="s", linestyle="-", label="Avg Time in microseconds (Age->Name)")
plt.xticks(range(10, 101, 10))
plt.xlabel("n (records per dataset)")
plt.ylabel("Average time in microseconds")
plt.title("Insertion Sort  Average time vs n")
plt.grid()
plt.legend()
plt.savefig("averagetime_plot.png", dpi=200)
```
