```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("comparisons.csv")
```

```python

plt.figure(figsize=(8,5))
plt.plot(df["n"], df["avg_age"],  marker="o", linestyle="-", label="Avg comparisons (Age)")
plt.plot(df["n"], df["avg_name"], marker="s", linestyle="-", label="Avg comparisons (Name)")
plt.plot(df["n"], df["avg_name_then_age"], marker="^", linestyle="-", label="Avg comparisons (NameAge)")

plt.xlabel("n (records per file)")
plt.ylabel("Average number of comparisons")
plt.title("Insertion Sort  Average Comparisons vs n")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.savefig("avg_comparisons.png", dpi=200)
print("Saved plot to avg_comparisons.png")
```
